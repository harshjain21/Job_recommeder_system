```python
data = pd.read_csv('/content/job_descriptions.csv', on_bad_lines= 'skip', engine = 'python')
jobs_df = data.copy()
```

```python
jobs_df.columns = jobs_df.columns.str.strip()
columns = jobs_df.columns.tolist()

new_columns = [col.replace(' ', '_') for col in columns]
jobs_df.columns = new_columns
jobs_df.sample(10)
```

```python
# --- Sampling the jobs_df ---
desired_sample_size = 100000
actual_sample_size = min(desired_sample_size, jobs_df.shape[0])

# Columns to keep for the sampled DataFrame
selected_cols = ['Role', 'Job_Description', 'skills', 'Responsibilities', 'Job_Title']
if 'Category' in jobs_df.columns:
    selected_cols.append('Category')

# Perform random sampling and select columns
jobs_df_sampled = jobs_df.sample(n=actual_sample_size, random_state=50)[selected_cols].reset_index(drop=True)

print(f"Sampled jobs_df_sampled shape: {jobs_df_sampled.shape}")
print("\nSampled jobs_df_sampled head:")
jobs_df_sampled.head()
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
import warnings

warnings.filterwarnings('ignore')

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if pd.isna(text): # Handle NaN values
        return ""
    text = str(text).lower() # Convert to string and lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove punctuation and special characters
    text = re.sub(r'\d+', '', text) # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace

    # Remove stop words
    text = " ".join([word for word in text.split() if word not in stop_words])

    # Lemmatize
    text = " ".join([lemmatizer.lemmatize(word) for word in text.split()])
    return text

print("Text cleaning function defined.")


# List of relevant text columns from job listings
job_text_cols = ['Job_Title', 'Role', 'Job_Description', 'skills', 'Responsibilities']

# Ensure columns exist and fill NaNs
for col in job_text_cols:
    if col not in jobs_df_sampled.columns:
        print(f"Warning: Job column '{col}' not found. Skipping or adjust column list.")
        # Remove missing columns from the list to avoid errors later
        job_text_cols.remove(col)
    else:
        jobs_df_sampled[col] = jobs_df_sampled[col].fillna('')

# Combine job text columns into a single 'full_job_text'
jobs_df_sampled['full_job_text'] = jobs_df_sampled[job_text_cols].agg(' '.join, axis=1)

# Apply cleaning
print("\nCleaning job listing text...")
jobs_df_sampled['cleaned_full_job_text'] = jobs_df_sampled['full_job_text'].apply(clean_text)
print("Job listing text cleaned.")
print("Sample of cleaned_full_job_text (first entry):")
print(jobs_df_sampled['cleaned_full_job_text'].head(1).iloc[0][:500], "...")

# List of relevant text columns from user profiles
user_text_cols = ['user_skills', 'user_educational_qualifications', 'user_work_experience', 'user_interests']

# Ensure columns exist and fill NaNs
for col in user_text_cols:
    if col not in user_df.columns:
        print(f"Warning: User profile column '{col}' not found. Skipping or adjust column list.")
        user_text_cols.remove(col)
    else:
        user_df[col] = user_df[col].fillna('')

# Combine user profile text columns into a single 'full_user_text'
user_df['full_user_text'] = user_df[user_text_cols].agg(' '.join, axis=1)

# Apply cleaning
print("\nCleaning user profile text...")
user_df['cleaned_full_user_text'] = user_df['full_user_text'].apply(clean_text)
print("User profile text cleaned.")
print("Sample of cleaned_full_user_text (first entry):")
print(user_df['cleaned_full_user_text'].head(1).iloc[0][:500], "...")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

print("Loading SentenceTransformer model ('all-MiniLM-L6-v2')...")
sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")

print("\nGenerating embeddings for job listings (cleaned_full_job_text)...")
job_texts = jobs_df_sampled['cleaned_full_job_text'].tolist()
job_embeddings = sbert_model.encode(job_texts, show_progress_bar=True)
print(f"Generated {len(job_embeddings)} job embeddings of dimension {job_embeddings.shape[1]}.")

print("\nGenerating embeddings for user profiles (cleaned_full_user_text)...")
user_texts = user_df['cleaned_full_user_text'].tolist()
user_embeddings = sbert_model.encode(user_texts, show_progress_bar=True)
print(f"Generated {len(user_embeddings)} user embeddings of dimension {user_embeddings.shape[1]}.")

print("\nText data vectorized!")


'''Now that we have both `user_embeddings` and `job_embeddings`, we can calculate the similarity scores'''

# This function will calculate similarity for ONE user against ALL jobs
def get_job_recommendations_for_user(user_embedding, all_job_embeddings, jobs_df_sampled, top_n=10):
    """
    Calculates cosine similarity between a single user's embedding and all job embeddings,
    and returns the top N recommended jobs.

    Args:
        user_embedding (np.array): The embedding vector for a single user.
        all_job_embeddings (np.array): A 2D array of all job embedding vectors.
        jobs_df (pd.DataFrame): The DataFrame containing job details.
        top_n (int): The number of top recommendations to return.

    Returns:
        pd.DataFrame: A DataFrame of top N recommended jobs with their similarity scores.
    """
    # Reshape user_embedding to be 2D for cosine_similarity function
    user_embedding_reshaped = user_embedding.reshape(1, -1)

    # Calculate cosine similarities between this user and all jobs
    # The output will be a 2D array, so we flatten it to a 1D array
    similarities = cosine_similarity(user_embedding_reshaped, all_job_embeddings).flatten()

    # Get the indices of jobs sorted by similarity in descending order
    top_job_indices = similarities.argsort()[-top_n:][::-1]

    # Get the top N jobs and their scores
    recommended_jobs = jobs_df_sampled.iloc[top_job_indices].copy() # .copy() to avoid SettingWithCopyWarning
    recommended_jobs['similarity_score'] = similarities[top_job_indices]

    return recommended_jobs

print("\nCosine similarity calculation function defined.")

# --- Example: Recommendations for the first dummy user ---
example_user_id = 1
example_user_index = example_user_id - 1 # User_id starts from 1 and matches DataFrame index + 1
target_user_embedding = user_embeddings[example_user_index]

print(f"\n--- Top 10 Job Recommendations for User ID: {example_user_id} ({user_df.loc[example_user_index, 'name']}) ---")
print(f"User Skills: {user_df.loc[example_user_index, 'user_skills']}")
print(f"User Interests: {user_df.loc[example_user_index, 'user_interests']}")
print("-" * 60)

top_recommendations = get_job_recommendations_for_user(
    target_user_embedding,
    job_embeddings,
    jobs_df_sampled,
    top_n=5 # Showing top 5 for brevity in output
)

# Display relevant columns of the recommended jobs
print(top_recommendations[['Job_Title', 'skills', 'similarity_score']])
```

```python
print("\n--- Enter Custom User Profile Details ---")
print("Provide details as if it were a short resume summary or LinkedIn profile.")
custom_skills = input("Enter your skills: ")
custom_education = input("Enter your educational qualifications : ")
custom_experience_details = input("Enter your work experience details : ")
custom_interests = input("Enter your job interests/preferred fields: ")

# Combine all custom inputs into a single string for cleaning and embedding
custom_user_text_raw = f"{custom_skills} {custom_education} {custom_experience_details} {custom_interests}"
cleaned_custom_user_text = clean_text(custom_user_text_raw)

print(f"\nCleaned Custom User Text for Embedding: {cleaned_custom_user_text[:200]}...") # Show a snippet

# --- Generate Embedding for Custom Input ---
print("\nGenerating embedding for custom user profile...")
custom_user_embedding = sbert_model.encode([cleaned_custom_user_text], show_progress_bar=False)[0]
print("Custom user embedding generated.")

# --- Get and Display Recommendations ---
top_n_recommendations = 10
print(f"\n--- Top {top_n_recommendations} Recommended Job Postings ---")

custom_recommendations_df = get_job_recommendations_for_user(
    custom_user_embedding,
    job_embeddings, # Using the embeddings for the sampled jobs
    jobs_df_sampled, # Using the sampled DataFrame for details
    top_n=top_n_recommendations
)


display_cols = ['Job_Title', 'skills', 'similarity_score']
if 'Category' in custom_recommendations_df.columns:
    display_cols.insert(1, 'Category') # Insert Category after Job_Title for better readability

print(custom_recommendations_df[display_cols])

print("\nCustom recommendations generated and displayed!")
```
